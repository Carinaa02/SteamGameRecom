```python
from sklearn.model_selection import train_test_split
import numpy as np
import random
```

```python
# Randomly split into train (70%), validation (15%), test (15%)
train_df, temp_df = train_test_split(recommendations_df, test_size=0.3, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)
```

```python
train_df['hybrid_score'] = (
    0.5 * train_df['is_recommended'] +
    0.4 * train_df['hours'] +
    0.1 * train_df['helpful']
)

game_scores = train_df.groupby('app_id').agg({
    'hybrid_score': 'mean',
    'user_id': 'count'  # num reviews
}).rename(columns={'user_id': 'num_reviews'}).reset_index()
```

```python
hybrid_df['final_score'] = (
    0.5 * hybrid_df['hybrid_score'] +
    0.4 * hybrid_df['rating'] -
    0.1 * hybrid_df['price_final']
)
```

```python
top_games = hybrid_df.sort_values('final_score', ascending=False).head(10)
top_games
```

```python
def recommend_for_user(user_id, train_df, hybrid_df, top_n=10):
    seen = train_df[train_df['user_id'] == user_id]['app_id'].unique()
    unseen = hybrid_df[~hybrid_df['app_id'].isin(seen)]
    return unseen.sort_values('final_score', ascending=False).head(top_n)
```

```python
def get_user_recommendation_metrics(user_id, test_df, train_df, hybrid_df, k=10):
    # Get actual relevant games
    actual = test_df[
        (test_df['user_id'] == user_id) &
        (test_df['is_recommended'] == 1)
    ]['app_id'].unique()

    if len(actual) == 0:
        return None  # skip users with no ground truth positives

    # Get recommended games
    recommended = recommend_for_user(user_id, train_df, hybrid_df, top_n=k)['app_id'].values

    # Precision, Recall
    intersection = set(actual) & set(recommended)
    precision = len(intersection) / k
    recall = len(intersection) / len(actual)
    
    # F1
    if precision + recall > 0:
        f1 = 2 * (precision * recall) / (precision + recall)
    else:
        f1 = 0.0

    # NDCG
    dcg = 0
    for i, app in enumerate(recommended):
        if app in actual:
            dcg += 1 / np.log2(i + 2)  # +2 because i starts from 0

    # Ideal DCG
    ideal_dcg = sum([1 / np.log2(i + 2) for i in range(min(len(actual), k))])
    ndcg = dcg / ideal_dcg if ideal_dcg > 0 else 0.0

    return precision, recall, f1, ndcg
```

```python
all_users = test_df['user_id'].unique()
sample_users = random.sample(list(all_users), 10000)
```

```python
train_sample = train_df[train_df['user_id'].isin(sample_users)]
test_sample = test_df[test_df['user_id'].isin(sample_users)]
```

```python
metrics = []

for user_id in sample_users:
    result = get_user_recommendation_metrics(user_id, test_sample, train_sample, hybrid_df, k=5)
    if result:
        metrics.append(result)

if metrics:
    precision, recall, f1, ndcg = zip(*metrics)
    print("Precision@10:", sum(precision)/len(precision))
    print("Recall@10:", sum(recall)/len(recall))
    print("F1@10:", sum(f1)/len(f1))
    print("NDCG@10:", sum(ndcg)/len(ndcg))
    print("Users Evaluated:", len(metrics))
else:
    print("No valid users to evaluate.")
```
